# 5 bibliotecas para coletar dados do mercado financeiro

## 1. Yahoo Finance
- Dados da cotação de ações brasileiras e estrangeiras
- Balanço patrimonial das empresas e o fluxo de caixa
- Informações estatísticas e distribuição acionárias
- Cotação do dolar e indices

```python
import yfinance as yf
```

```python
# Definir um ticker
tickerSymbol = 'TAEE11.SA'
tickerData = yf.Ticker(tickerSymbol)
```

```python
tickerData
```

```python
# consultar os dados históricos
tickerDf = tickerData.history(interval='1d',
                              start='2010-01-01',
                              end='2021-01-01')
```

```python
tickerDf
```

## 3. Quandl
- Dados financeiros, econômicos e alternativos 
- **link:** https://www.quandl.com/

```python
import quandl
```

```python
f = open('quandl_api.txt', 'r')
api_key = f.read()
```

```python
data = quandl.get("BSE/BOM500180", authtoken=api_key)
data
```

```python
data = quandl.get_table('WIKI/PRICES', ticker = ['AAPL', 'MSFT'], 
                        qopts = { 'columns': ['ticker', 'date', 'adj_close'] }, 
                        date = { 'gte': '2015-12-31', 'lte': '2016-12-31' }, 
                        paginate=True)
```

```python
data
```

## 4. Pandas Reader
- Extrair recursos da internet para um pandas dataframe
- **link**: https://pandas-datareader.readthedocs.io/en/latest/remote_data.html
- Resources:
    - Yahoo Finance
    - IEX
    - Alpha Vantage
    - Stooq

```python
from pandas_datareader.famafrench import get_available_datasets
from pandas_datareader import data
```

```python
datasets = get_available_datasets()
```

```python
datasets
```

```python
tickers = 'AAPL'

start_date = '2010-01-01'
end_date = '2021-01-01'

dados = data.DataReader(tickers, 'yahoo', start_date, end_date)
```

```python
dados
```
